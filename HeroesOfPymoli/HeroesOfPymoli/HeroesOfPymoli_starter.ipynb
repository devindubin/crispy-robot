## Age Demographics

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

```python
age_bins = pd.IntervalIndex.from_tuples([(0,9),(9,14),(14,19),(19,24),(24,29),(29,34),(34,39),(39,50)])
ages = pd.cut(dropped_d['Age'],age_bins)
age_df =pd.DataFrame({'Total Count':ages.value_counts(),
                      'Percentage of Players':round(100*(ages.value_counts()/ages.value_counts().sum()),2)})
age_df.reset_index(drop = False)
# sorted_age = age_df.sort_index()
# sorted_age
for perc in range(len(age_df['Percentage of Players'])):
    age_df.iloc[perc,1] = "{:.2f}%".format(age_df.iloc[perc,1])
age_df = age_df.sort_index()
age_df.set_index([['<10',"10-14","15-19","20-24","25-29","30-34","35-39","40+"]])
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender_data = purchase_data.copy().copy()
top_spend_SN_group = top_spender_data.groupby('SN')
top_5_players = list(top_spend_SN_group['Price'].sum().sort_values(ascending=False).head(5).keys())
#top_5_pcount = top_spend_SN_group[top_spend_SN_group['SN'] == 'Lisosia93']['Purchase ID'].count()
#top_spend_SN_group_df[[top_spend_SN_group_df['SN'] == 'Lisosia93']].count()
top_5_players

#Trying to form a group of only the players with the top 5 highest spending
top_spend_top5_filtered = top_spender_data.groupby('SN').filter(lambda x: x['SN'].all() in top_5_players)
top_spend_top5_group = top_spend_top5_filtered.groupby('SN')
ts_pc_s = top_spend_top5_group['Price'].count()
ts_tpv_s = top_spend_top5_group['Price'].sum()
ts_app_s = ts_tpv_s/ts_pc_s

# print(ts_pc_s)
# print(ts_tpv_s)
# print(ts_app_s)


summary_ts = pd.DataFrame({'Purchase Count':ts_pc_s,
                          'Average Purchase Price':ts_app_s,
                          'Total Purchase Value':ts_tpv_s})

for r in range(len(top_5_players)):
    for c in range(1,3):
        summary_ts.iloc[r,c] = "${:.2f}".format(round(summary_ts.iloc[r,c],2))
        summary_ts.iloc[r,c]
summary_ts = summary_ts.sort_values('Total Purchase Value',ascending = False)
summary_ts
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
mpt = purchase_data[['Item ID','Item Name','Price']]
item_id_name_group = mpt.groupby(['Item ID','Item Name'])
mpt_pc = item_id_name_group['Item ID'].count()
mpt_ip = item_id_name_group['Price'].mean()
mpt_tpv = item_id_name_group['Price'].sum()

summary_mpd = pd.DataFrame({'Purchase Count':mpt_pc,
                           'Item Price':mpt_ip})


summary_mpt = pd.merge(summary_mpd,mpt_tpv,how = "inner",on = ['Item ID','Item Name'])
summary_mpt = summary_mpt.rename(columns = {'Price' : 'Total Purchase Value'})
summary_mpt = summary_mpt.sort_values(by = ['Purchase Count'],ascending = False)
summary_mpi = mpt_tpv
for r in range(len(summary_mpt)):
    for c in range(1,3):
        summary_mpt.iloc[r,c] = "${:.2f}".format(round(summary_mpt.iloc[r,c],2))
        
summary_mpt.head()
```
